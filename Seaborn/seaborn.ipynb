# Categorical Plots

## Bar plots

```python
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.cov)
#estimator - estimate  data based on mean,median,std,var,covariance(cov)
```

## Count Plot

```python
sns.countplot(x='sex',data=tips_df)
```

## Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')

#Note 
# 1.the middle line is the median
# 2.box end point is 1 standard deviation from the median
# 3.the "DOTS" outside the box are outliers in the data
# 4.whiskers - the data after 1 standard deviation (|---[ | ]--|)
# 5.hue -another category for the plot


#repositioning of Legend of graph
plt.legend(loc=0)

#can use both Location String or location Code in plt.legend(loc=?)

# Location String	Location Code
# 'best'	0
# 'upper right'	1
# 'upper left'	2
# 'lower left'	3
# 'lower right'	4
# 'right'	5
# 'center left'	6
# 'center right'	7
# 'lower center'	8
# 'upper center'	9
# 'center'	10
```

## Violin Plot
### combination of box and KDE plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=False)
#split comparing categories to each other

plt.legend(loc=0)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
#split comparing categories to each other

plt.legend(loc=0)
```

## Strip Plot

```python
#stip plot is scatter plot adding 1 more column as category
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=False)
#jitter - spreading data points in the plot to avoid any stacked points
#dodge - split plot on hue 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
#dodge - split plot on hue 
```

## Swarm Plot

# PairGrid

```python
#creating scatter plots in Grid 
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map(plt.scatter)
```

```python
#plotting histogram and scatter in diagonal
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
#plotting histogram in diagonal
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
#plotting scatter and kde plot in upper and lower matrix concept
# and adding histogram in diagnal
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

##### custom plot

```python
iris_g = sns.PairGrid(iris , hue='species',
                     x_vars = ['sepal_length','sepal_width'],
                     y_vars = ['petal_length','petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

### Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df,col='time',row='smoker')
#creating the row and column figure

tips_fg.map(plt.scatter,'total_bill','tip')
#plotting facet grid
```

```python
tips_fg = sns.FacetGrid(tips_df,col='time',hue='smoker',height=4,aspect=1.3
                       ,col_order=['Dinner','Lunch'],palette='Set1')
tips_fg.map(plt.scatter,'total_bill','tip',edgecolor='w')
```

```python
#creating differenet markers of different size
kws = dict(s=50,linewidth=0.5,edgecolor='w')
tips_fg= sns.FacetGrid(tips_df,col='sex',hue='smoker',height=4,aspect=1.3,
                      hue_order=['Yes','No'],
                      hue_kws = dict(marker=['^','v']))
tips_fg.map(plt.scatter,'total_bill','tip',**kws)
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df,col='subject',col_wrap=5,height=1.5)
att_fg.map(plt.plot,'solutions','score',marker='*')
```
